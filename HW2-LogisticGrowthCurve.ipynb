## Question d: Write a Python program that starts with an initial population and solves the model equation as a function of time. This can be done by modifying the logistic growth model written in class. 

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib notebook
```

```python
#define parameters
No = 5000000      #initial population [tons]

K = 10000000 #carrying capacity [tons]
r = 0.4      #maximum per capita growth rate [1/yr]

#add fixed quota
ECN = 100000 # annual catch [tons/year]

#additional parameters
E = 100 #effort [# boats]
C = .001 #catchablity constant [tons/boat]
#C doubles from - .001 to .002 for part e of HW

#define time parameters
Ti = 0        #initial time [yr]
Tf = 30       #final time [yr]
delta_T = 1   #timestep [yr]
```

```python
#Create an array containing each time the model will be evaluated, and an array for the population at each time 
#(since we're breaking a continuous logistic growth model to determine population at discrete times)

t = np.arange(Ti,Tf,delta_T) #initial time, final time, time step
nsteps = len(t) #number of timesteps in model

N = np.zeros(nsteps)
N[0] = No
```

```python
#calculate population for each timestep 
for i in range(0,nsteps-1):
  dNdt = r*N[i]*((1-N[i])/K) - E*C*N[i]
  N[i+1] = N[i]+ dNdt*delta_T
```

```python
print(N)
```

```python
plt.figure()
plt.plot(t,N,'-o')
plt.title('Catchability as a function of population')
plt.xlabel('time [years]')
plt.ylabel('population [tons of fish]');
#plt.text(2,-.4e8,'c=.001, ',size=15)
```
